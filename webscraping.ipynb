```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.checkraka.com/car/?quicksearch_order=update%2CDESC&page=2'
```

```python
data = requests.get(url)
data
```

```python
soup = BeautifulSoup(data.text)
```

```python
car = soup.find('div',{'class':'content c158'})
```

```python
name = car.find('div',{'class':'data'}).find('a').text
name
```

```python
price = car.find('div',{'class':'price'}).text
price
```

```python
len(soup.find_all('div',{'class':'content c158'}))
```

```python
name_ = []
price_ = []
page = 1
for i in range(95):
  url = "https://www.checkraka.com/car/?quicksearch_order=update%2CDESC&page=" + str(page)
  data = requests.get(url)
  soup = BeautifulSoup(data.text)
  print(url)
  page += 1
  for i in soup.find_all('div',{'class':'content c158'}):
      name_.append(i.find('div',{'class':'data'}).find('a').text)
      price_.append(i.find('div',{'class':'price'}).text)
```

```python
url
```

```python
for i in soup.find_all('div',{'class':'content c158'}):
  name_.append(i.find('div',{'class':'data'}).find('a').text)
  price_.append(i.find('div',{'class':'price'}).text)
```

```python
name_
```

```python
price_
```

```python
import pandas as pd
```

```python
table = pd.DataFrame([name_,price_]).transpose()
table
```

```python
table.columns = ['carname','price']
table
```

```python
table.set_index('carname')
table
```

```python
table.to_csv('cartable.csv')
```
